## **Step 6: Searching and downloading scientific literatures on the query term.**

- -q = query (for example: Ocean Heat, it can be changed to any keyword)
- -k = number of papers (Here it is 10, this can be any number)
- -o = output file name (Here it is Ocean_Heat2, this can be any name)
- -x = paper in XML format
- -p = paper in pdf format
- --fulltext_html = paper in HTML format
- --datatables = create datatables

**See more options from GitHub repository : [pygetpapers](https://github.com/petermr/pygetpapers)**

```python
!pygetpapers -q "Ocean heat" -k 10 -o Ocean_Heat2 -x -p --fulltext_html --datatables
```

## **Step 7: Downloading the results and saving them in local machine.**

```python
import shutil
from google.colab import files

# Step 1: Zip the entire folder
shutil.make_archive('/content/pygetpapers/Ocean_Heat2', 'zip', '/content/pygetpapers/Ocean_Heat2')

# Step 2: Download the zipped folder
files.download('/content/pygetpapers/Ocean_Heat2.zip')
```
